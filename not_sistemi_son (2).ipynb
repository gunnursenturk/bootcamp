```python
import pandas as pd
```

```python
ogrenci_adlar = []
ogrenci_soyadlar = []
ogrenci_nolar = []
ogrenci_vizeler_1 = []
ogrenci_vizeler_2 = []
ogrenci_finaller = []
ogrenci_ortalamalar = []
ogrenci_harf_notlar = []
ogrenci_durumlar =[]

data = {
    'Ad': ogrenci_adlar,
    'Soyad': ogrenci_soyadlar,
    'No': ogrenci_nolar,
    'Vize1': ogrenci_vizeler_1,
    'Vize2': ogrenci_vizeler_2,
    'Final': ogrenci_finaller,
    'Ortalama': ogrenci_ortalamalar,
    'HarfNotu': ogrenci_harf_notlar,
    'Durum': ogrenci_durumlar,
}
```

```python
class Ogrenci:
    def __init__(self, ad = 'Null', soyad = 'Null', no = 0, vize1 = 0, vize2 = 0, final = 0, ortalama = 0, harf_notu = '', durum = 'Hesaplanmadı' ):
        self.ad = ad
        self.soyad = soyad
        self.no = no
        self.vize1 = vize1
        self.vize2 = vize2
        self.final = final
        self.durum = durum
        self.ortalama = ortalama
        self.harf_notu = harf_notu
        
# Öğrenci Kayıt Fonksiyonu:
    def ogrenci_kayit(self):
        ad = input('Lütfen öğrenci adını giriniz: ')
        self.ad = ad
        soyad = input('Lütfen öğrenci soyadını giriniz: ')
        self.soyad = soyad
        no = int(input('Lütfen öğrenci no giriniz: '))
        self.no = no
        vize1 = float(input('Lütfen 1. vize sınavı notunu giriniz: '))
        self.vize1 = vize1
        vize2 = float(input('Lütfen 2. vize sınavı notunu giriniz: '))
        self.vize2 = vize2
        final = float(input('Lütfen final sınavı notunu giriniz: '))
        self.final = final
        
# Ortalama Hesaplama:        
        _vize1 = self.vize1
        _vize2 = self.vize2
        _final = self.final
        
        self.ortalama = (_vize1 * 0.25) + (_vize2 * 0.25) + (_final * 0.5)
        
# Harf Notu Hesaplama:
        
        if(100 >= self.ortalama >= 80):
          self.harf_notu = 'AA'
        elif(80 > self.ortalama >= 70):
          self.harf_notu = 'BB'
        elif(70 > self.ortalama >= 60):
          self.harf_notu = 'CC'
        elif(60 > self.ortalama >= 50):
          self.harf_notu = 'DD'
        else: 
          self.harf_notu = 'FF'
        
# Durum Hesaplama:        
        if(self.harf_notu == 'FF'):
            self.durum = 'KALDI'
        else:
            self.durum = 'GEÇTİ'

# Kayıt Özeti:
        print(f"""
            Öğrenci bilgi girişi yapıldı;
            Öğrenci Adı: {self.ad}
            Öğrenci Soyadı: {self.soyad}
            Öğrenci No: {self.no}
            1. Vize Notu: {self.vize1}
            2. Vize Notu: {self.vize2}
            Final Notu: {self.final}
            Ortalama Notu: {self.ortalama}
            Harf Notu: {self.harf_notu}
            Öğrenci Durumu = {self.durum}
            """)
```

```python
def ogrenci_ekle(ogrenci):
    
    _ogrenci = ogrenci
    
    ogrenci_adlar.append(_ogrenci.ad)
    ogrenci_soyadlar.append(_ogrenci.soyad)
    ogrenci_nolar.append(_ogrenci.no)
    ogrenci_vizeler_1.append(_ogrenci.vize1)
    ogrenci_vizeler_2.append(_ogrenci.vize2)
    ogrenci_finaller.append(_ogrenci.final)
    ogrenci_ortalamalar.append(_ogrenci.ortalama)
    ogrenci_harf_notlar.append(_ogrenci.harf_notu)
    ogrenci_durumlar.append(_ogrenci.durum)
    
    print('Öğrenci kayıt işlemi başarıyla gerçekleştirildi!')
```

```python
ogrenci_n = Ogrenci()
```

```python
n = 0
while n < 10: #Sınıf nüfusu tanımlanabilir.
  ogrenci_n.ogrenci_kayit()
  ogrenci_ekle(ogrenci_n)
  n +=1
 
```
